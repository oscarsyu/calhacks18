```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
spotify_data = pd.read_csv("calhacksdata2clean.csv")
```

```python
#datacleaning
```

```python
spotify_data = spotify_data.drop_duplicates(subset='title', keep = 'first')
```

```python
spotify_data  = spotify_data.drop(['top genre', 'artist', 'year', 'added', 'live', 'pop'], axis = 1)
```

```python
zero_bpm = spotify_data[spotify_data["bpm"] == 0].index[0]
spotify_data = spotify_data.drop([zero_bpm])
```

```python
train, test = train_test_split(spotify_data)
```

```python
train
```

```python
min_val = train["val"].sort_values().iloc[0]
max_val = train["val"].sort_values(ascending=False).iloc[0]
```

```python
val_range = max_val - min_val
bins = val_range/5
```

```python
moods = np.arange(min_val, max_val, bins)
```

```python
def mood_convert(mood):
    if mood < moods[0]:
        return 0
    if moods[0] <= mood < moods[1]:
        return 1
    if moods[1] <= mood < moods[2]:
        return 2
    if moods[2] <= mood < moods[3]:
        return 3
    if mood >= moods[3]:
        return 4
```

```python
train["mood"] = train["val"].apply(lambda x: mood_convert(x))
test["mood"] = test["val"].apply(lambda x: mood_convert(x))
```

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
mod = LogisticRegressionCV()
```

```python
dropped = train.drop(['title', 'mood'], axis = 1)
```

```python
mod.fit(dropped, train["mood"])
```

```python
mod.predict(train.iloc[:, 1:9])
```

```python
mod.score(train.iloc[:, 1:9], train["mood"])
```

```python
test.iloc[:, 1:9]
```

```python
mod.score(test.iloc[:, 1:9], test["mood"])
```

```python
result = mod.predict(test.iloc[:, 1:9])
```

```python
test["predicted"] = result
```

```python
bpm_range = max(train["bpm"])- min(train["bpm"])
energy_range = max(train["nrgy"]) - min(train["nrgy"])
dnce_range = max(train["dnce"]) - min(train["dnce"])
val_range = max(train["val"]) - min(train["val"])
acous_range = max(train["acous"]) - min(train["acous"])
```

```python
train_norm = train
train_norm["bpm"] = train['bpm'] / bpm_range
train_norm["nrgy"] = train["nrgy"] / energy_range
train_norm["dnce"] = train["dnce"] / dnce_range 
train_norm["val"] = train["val"] / val_range
train_norm["acous"] = train["acous"] / acous_range
```
